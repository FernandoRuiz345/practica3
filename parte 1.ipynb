## Realice la conexión mediante requests y verifique que la respuesta HTTP sea exitosa. Muestre el status_code y explique su significado.

La coneccion al sitio web fue exitosa.

```python
import requests

url = "https://documenter.getpostman.com/view/5578104/RWgqUxxh"

respuesta = requests.get(url)

# Mostrar el código de estado
print("Status Code:", respuesta.status_code)

# Verificar si la respuesta fue exitosa
if respuesta.status_code == 200:
    print("La conexión fue exitosa.")
else:
    print("ERROR.")
```

```python
import requests

url = "https://documenter.getpostman.com/view/5578104/RWgqUxxh"

respuesta = requests.get(url)

print("URL consultada:", respuesta.url)
print("Content-Type:", respuesta.headers.get("Content-Type"))
```
